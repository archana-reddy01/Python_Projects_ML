```python
#importing dependencies
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#data collection and Analysis
dias_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
dias_dataset.head()
```

```python
dias_dataset.shape
```

```python
dias_dataset.describe()
```

```python
dias_dataset['Outcome'].value_counts()
```

0 -- >Non- Diabetic

1 -- > Diabetic

```python
dias_dataset.groupby('Outcome').mean()
```

```python
#seperating data and labels
X=dias_dataset.drop(columns='Outcome',axis=1)
Y=dias_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
#we can use scaler.fit_transform(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X=standardized_data
Y=dias_dataset['Outcome']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=3)
```

```python
print(X_train)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine classifier
classifier.fit(X_train,Y_train)
```

```python
#model Evaluation
#Accuracy sore
X_train_prediction=classifier.predict(X_train)
accuracy_train=accuracy_score(X_train_prediction,Y_train)
```

```python
print("The Accuracy of trainning data : ",accuracy_train)
```

```python
X_test_prediction=classifier.predict(X_test)
accuracy_test=accuracy_score(X_test_prediction,Y_test)
```

```python
print("The Accuracy of test data : ",accuracy_test)
```

```python
#making a predictive system
input_data = (3,78,50,32,88,31,0.248,26)

#changing the input data into numpy array for more efficiency and easy
input_to_numpy =np.asarray(input_data)

#reshape the array as we are using for one instance
input_reshaped=input_to_numpy.reshape(1,-1)

#standardizind the data because we alreay modified our data  i.e standardized
standard_data = scaler.transform(input_reshaped)
print(standard_data)
prediction=classifier.predict(standard_data)
print(prediction)

if prediction[0]==1:
  print("The person have diabetes")
else:
  print("The person is Non-Diabetic")
```
