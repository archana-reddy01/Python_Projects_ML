```python
# Handling missing values
# mean-->average
# mode --> most repeated value
big_mart_data['Item_Weight'].mean()
```

```python
#filling missing values in 'Item_Weight' column with mean value
big_mart_data['Item_Weight'].fillna(big_mart_data['Item_Weight'].mean(),inplace=True)
```

```python
big_mart_data.isnull().sum()
```

replacing the missing values in 'Outlet_size' with mode


```python

mode_of_outlet = big_mart_data.pivot_table(values='Outlet_Size',columns='Outlet_Type',aggfunc=(lambda x:x.mode()[0]))
```

```python
print(mode_of_outlet)
```

```python
missing_values = big_mart_data['Outlet_Size'].isnull()
```

```python
print(missing_values)
#false --> value is present True --> value is  missing
```

```python

#to replce the missing values with mode
big_mart_data.loc[missing_values,'Outlet_Size']=big_mart_data.loc[missing_values,'Outlet_Type'].apply(lambda x:mode_of_outlet[x].values[0])
```

Data preprocessing

```python
big_mart_data.head()
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

```python
big_mart_data.replace({'Item_Fat_Content':{'low fat':'Low Fat','LF':'Low Fat','reg':'Regular'}},inplace=True)
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

```python
encoder = LabelEncoder()
```

```python
big_mart_data['Item_Identifier']=encoder.fit_transform(big_mart_data['Item_Identifier'])
big_mart_data['Item_Fat_Content']=encoder.fit_transform(big_mart_data['Item_Fat_Content'])
big_mart_data['Item_Type']=encoder.fit_transform(big_mart_data['Item_Type'])
big_mart_data['Outlet_Identifier']=encoder.fit_transform(big_mart_data['Outlet_Identifier'])
big_mart_data['Outlet_Size']=encoder.fit_transform(big_mart_data['Outlet_Size'])
big_mart_data['Outlet_Location_Type']=encoder.fit_transform(big_mart_data['Outlet_Location_Type'])
big_mart_data['Outlet_Type']=encoder.fit_transform(big_mart_data['Outlet_Type'])
```

```python
big_mart_data.head()
```

```python
x=big_mart_data.drop(columns='Item_Outlet_Sales',axis=1)
y=big_mart_data['Item_Outlet_Sales']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Training the model

```python
#Training model
reg =XGBRegressor()
```

```python
reg.fit(x_train,y_train)
```

```python
# prediction on training data
train_data_predictio = reg.predict(x_train)
```

```python
# R squared value
r2_train = metrics.r2_score(y_train,train_data_predictio)
```

```python
print("R squared value : ",r2_train)
```

```python
# testing data prediction
test_data_prediction = reg.predict(x_test)
```

```python
r2_test=metrics.r2_score(y_test,test_data_prediction)
```

```python
print("R squared value : ",r2_test)
```
